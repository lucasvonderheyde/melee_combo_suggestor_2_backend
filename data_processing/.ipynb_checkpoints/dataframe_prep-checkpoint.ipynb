```python
import pandas as pd
import json
from pandas import json_normalize
from google.cloud import storage
from dotenv import load_dotenv
import os

load_dotenv()

storage_client = storage.Client()

bucket_name = 'melee-combo-suggestor-2-slippi-storage'    
    
blob_higherportPostFrames = 'higherportPostFrames_442e7181-6d8b-4db7-b7a3-5b7e705a35bb.json'
blob_lowerportPostFrames = 'lowerportPostFrames_442e7181-6d8b-4db7-b7a3-5b7e705a35bb.json'

bucket = storage_client.bucket(bucket_name)
blob_higherport = bucket.blob(blob_higherportPostFrames)
blob_lowerport = bucket.blob(blob_lowerportPostFrames)

higherport_json = blob_higherport.download_as_text()

higherport_data = json.loads(higherport_json)

lowerport_json = blob_lowerport.download_as_text()

lowerport_data = json.loads(lowerport_json)

df1 = pd.DataFrame(pd.json_normalize(higherport_data))
df2 = pd.DataFrame(pd.json_normalize(lowerport_data))

df_higher_port_player_postframes = df1.add_prefix('higher_port_player_postframes_')
df_lower_port_player_postframes = df2.add_prefix('lower_port_player_postframes_')

df_higher_port_player_postframes_frame_column = df_higher_port_player_postframes.columns[0]
df_lower_port_player_postframes_frame_column = df_lower_port_player_postframes.columns[0]

combined_df = pd.merge(df_higher_port_player_postframes, df_lower_port_player_postframes, left_on=df_higher_port_player_postframes_frame_column, right_on=df_lower_port_player_postframes_frame_column, how='inner')

print(combined_df)


action_state_id_to_check_getting_hit = [0, 1, 2, 4, 8, 75, 76, 77, 78, 79, 80, 81, 82, 83, 84, 85, 86, 87, 88, 89, 90, 91, 227, 228, 239, 240, 241, 242]
death_action_state_ids = [0, 1, 2, 4, 8]

```
